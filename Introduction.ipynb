```python
# Import basic libraries
import pandas as pd

# Import Scikit-learn libraries for the dataset and model
from sklearn.datasets import load_iris
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

#For visualizing the confusion matrix
import matplotlib.pyplot as plt
import seaborn as sns

#Visualize model
from sklearn import tree
```

```python
iris = datasets.load_iris() # Load and prepare Dataset
print(iris.keys()) # Display the keys of the dataset
list(iris.target_names) # Display target names
```

```python
iris['feature_names'] # Display feature names(variable names)
```

```python

df = pd.DataFrame(data=iris.data, columns=iris.feature_names) # Define Data and columns
```

```python
df.head()
```

```python
# Create test / train 
X = df # Optional, but samples are often denoted as X
y = iris.target # What are our labels?

X_train, X_test, y_train, y_test = train_test_split(X, y) # Fill out function parameters

```

```python
#Create / fit model
model = DecisionTreeClassifier()# Define a model
model.fit(X_train, y_train) # Fit based on what?
```

```python
# Test model
pred = model.predict(X_test) # Predict on what?
print(X_test)
accuracy = accuracy_score(y_test, pred) # accuracy is calculated how?
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
#Plot results
cm =confusion_matrix(y_true=y_test, y_pred=pred)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True,fmt="d",  cmap = "Blues", xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
#Visualize model

plt.figure(figsize=(12,8))
tree.plot_tree(model, feature_names=iris.feature_names, class_names=iris.target_names, filled=False, rounded=True)
plt.show()
```
